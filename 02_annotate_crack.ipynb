```python
# Load your frames

import numpy as np
import napari

frames = np.load('processed/120nm/20240715_1934_51k_rescaled__2022_66k_merged_matched.npy', mmap_mode='r')
# Select frames to annotate with uniform spacing across the full sequence
# Target roughly 50 annotated frames regardless of sequence length

n_frames = frames.shape[0]
step = max(1, int(np.ceil(n_frames / 50)))
frames_to_annotate = list(range(0, n_frames, step))
frames_subset = frames[frames_to_annotate]

# Launch napari
viewer = napari.Viewer()
viewer.add_image(frames_subset, name='TEM frames', colormap='gray')
labels = viewer.add_labels(np.zeros(frames_subset.shape, dtype=np.uint8), name='crack_masks')

napari.run()  # Opens GUI window
```

```python
# Extract annotations from the napari labels layer
masks = labels.data

# Your data file name (without extension)
sample_name = "20240715_1934_51k_rescaled__2022_66k_merged_matched"

# Save with descriptive names to the annotations folder
np.save(f'annotations/120nm/{sample_name}_annotated_masks.npy', masks)
np.save(f'annotations/120nm/{sample_name}_frame_indices.npy', np.array(frames_to_annotate))
```
